# Limpando Base de Dados Simples

#### Esta base de dados é referente a clientes de um banco e sua situação de pagamento ou não pagamento de um empréstimo

## Importando bibliotecas e base de dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
base = pd.read_excel('ChavesClientes.xlsx')
base.head()
```

#### As colunas da base da dados parecem confusas no início, porém sua explicação é simples:
- A coluna "ChaveSituação" diz respeito a:
 * idade: 2 primeiros números 
 * sexo: M (masculino), F (feminino)
 * estado civil: S (solteiro), C (casado), D (divorciado), V (viuvo)
- A coluna "ClassRisco":
 * Classifica o cliente como A,B ou C e o indicador + ou -.
 * Cor do cliente de acordo com um modelo de churn interno da empresa ()
- A coluna "CatCliente" é formada por:
 * Categoria do cartão (Basic, Black, Platinum)
 * Categoria VIP (Alpha, Beta)
 
 Vamos começar simplesmente separando as colunas em colunas individuais

## Separando as colunas

```python
#Separando a ChaveSituacao
base['Idade'] = base.ChaveSituacao.apply(lambda string: string[:2])
base['Sexo'] = base.ChaveSituacao.apply(lambda string: string[2:3])
base['EstadoCivil'] = base.ChaveSituacao.apply(lambda string: string[-1])

#Separando a CatCliente
base['Categoria'] = base.CatCliente.str.split('-').str.get(0)
base['Cliente'] = base.CatCliente.str.split('-').str.get(1)

#Separando ClassRisco
base['Classe'] = base.ClassRisco.str[0]

def indicador(string):
    if '-' in string: return '-'
    elif '+' in string: return '+'
    else: return np.nan
    
base['Indicador'] = base.ClassRisco.apply(indicador)

def risco(risco):
    if 'cinza' in risco.lower(): return "cinza"
    elif 'amarelo' in risco.lower(): return 'amarelo'
    elif 'verde' in risco.lower(): return 'verde'
    elif 'roxo' in risco.lower(): return 'roxo'
    else: return np.nan
    
base['Risco'] = base.ClassRisco.apply(risco)

#Converteremos a coluna idade para o formato numérico, pois agora ela está como objeto

base['Idade'] = pd.to_numeric(base['Idade'])
```
